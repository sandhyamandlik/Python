```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5]
y=[1,4,9,16,25]

plt.plot(x,y)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title("Basic Line plot")
plt.show()
```

```python
x=[1,2,3,4,5]
y=[1,4,9,16,25]

## create a customized line plot

plt.plot(x,y,color='red',linestyle='--',marker='o',linewidth=3,markersize=9)
plt.grid(True)
```

```python
## multiple plots
## sample data

x=[1,2,3,4,5]
y1=[1,4,6,16,25]
y2=[1,2,3,4,5]

plt.figure(figsize=(9,5))

plt.subplot(1,2,1)
plt.plot(x,y1,color='green')
plt.title("plot 1")

plt.subplot(2,2,2)
plt.plot(y1,x,color='red')
plt.title("plot 2")

plt.subplot(2,2,3)
plt.plot(x,y2,color='pink')
plt.title("plot 3")

plt.subplot(2,2,4)
plt.plot(x,y2,color='blue')
plt.title("plot 4")
```

```python
## bar plot
categories=['A','B','C','D','E']
values=[5,6,7,8,6]

##create a bar plot
plt.bar(categories,values,color='purple')
plt.xlabel('categories')
plt.ylabel('values')
plt.title('Bar plot')
plt.show()
```

# Histogram

```python
# sample data

data=[1,2,2,3,3,4,4,4,5,5,6,6,6,6]

## create a histogram

plt.hist(data,bins=5,color='orange',edgecolor='black')
```

```python
## creating a scatter plot
## sample data
x=[1,2,3,4,5]
y=[2,3,4,5,6]

plt.scatter(x,y,color='blue',marker='x')
```

```python
import matplotlib.pyplot as plt

labels = ['A', 'B', 'C', 'D']
sizes = [30, 20, 40, 10]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.2, 0, 0, 0)  # This is for the explode effect, pulling out the first slice

# Create a pie chart
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct="%1.1f%%", shadow=True)

plt.show()
```

```python
## sales data visualization

import pandas as pd
sales_data_df=pd.read_csv('sales_data.csv')
(sales_data_df.head(5))
```

```python
## plot total  sales by product
total_sales_by_product=sales_data_df.groupby('category')['Total Sales'].sum()
print(total_sales_by_product)
```

```python
total_sales_by_product.plot(kind='bar',color='teal')
```

```python
## plot sales trend over time

sales_trend=sales_trend=sales_data_df.groupby('Date')['Total Sales'].sum().reset_index()
plt.plot(sales_trend['Date'],sales_trend['Total Sales'])
```
