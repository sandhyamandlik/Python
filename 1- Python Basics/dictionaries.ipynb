```python
# single key is always used
student={"name":"Krish","age":32,"name":24}
print(student)
```

```python
##accesing dictionary elemnets
student={"name":"Krish","age":32,"grade":'A'}
print(student)
```

```python
#accessing dictionary elements
print(student['grade'])
print(student['age'])

#accessing using get() method
print(student.get('grade'))
print(student.get('last_name'))
print(student.get('last_name',"not available"))
```

```python
## modifying dictionary elements
## dictionaries are mutable,so you can add ,update or delete elements

print(student)
```

```python
##update value for a key

student["age"]=32
print(student)
student["address"]="india" ## added a new key and value
print(student)
```

```python
del student['grade'] ## delete key and value pair
```

```python
## dictionary methods

keys=student.keys() 
print(keys)
values=student.values()
print(values)

items=student.items()
print(items)
```

```python
## shallow copy
student_copy=student
print(student)
print(student_copy)
```

```python
student["name"]="sandhya"
print(student)
print(student_copy)
```

```python
student_copy1=student.copy()# shallow copy
print(student_copy1)
print(student)
```

```python
##iterating over dictionaries
## you can use loops to iterate over dictionaries .keys,values,or items

## iterating over keys
for keys in student.keys():
  print(keys)
```

```python
## iterate over values
for value in student.values():
  print(value)
```

```python
## iterate over key value pairs
for key,value in student.items():
  print(f"{key}:{value}")
```

```python
## nested dictionaries
students={
"student1":{"name":"sandhya","age":32},
"student2":{"name":"sayli","age":35}
}

print(students)
```

```python
## access nested dictionaries elements
print(students["student2"]["name"])
print(students["student2"]["age"])
```

```python
## iterating over nested dictionaries
for student_id,student_info in students.items():
  print(f"{student_id}:{student_info}")
  for key,value in student_info.items():
    print(f"{key}:{value}")
```

```python
## Dictionaries comprehension
squares={x:x**2 for x in range(5)}
print(squares)
```

```python
## conditional dictionary comprehension
evens={x:x**2 for x in range(10) if x%2==0}
print(evens)
```

```python
## pratical examples
## use a dictionary to count the frequency of elements in list

numbers=[1,2,2,3,3,3,4,4,4,4]
frequency={}

for number in numbers:
  if number in frequency:
    frequency[number]+=1

  else:
    frequency[number]=1
print(frequency)
  
```

```python
##merge 2 dictionaries into one

dict1={"a":1,"b":2}
dict2={"b":3,"c":4}
merge_dict={**dict1,**dict2}
print(merge_dict)
```
