```python
## attributes of numpy

arr =np.array([[1, 2, 3], [4, 5, 6]])

print("Array:\n", arr)

print("Shape:", arr.shape) # Output: (2, 3)

print("Number of dimensions:", arr.ndim) # Output: 2

print("Size (number of elements):", arr.size) # Output: 6

print("Data type:", arr.dtype) # Output: int64 (may vary based on platform)

print("Item size (in bytes):", arr.itemsize) # Output: 8 (may vary based on platform)
```

```python
## numpy vectorized operation

arr1=np.array([1,2,3,4,5])
arr2=np.array([10,20,30,40,50])

## element wise addition,substraction,multiplication,division
print("addition:",arr1+arr2)
print("substraction:",arr1-arr2)
print("multiplication:",arr1*arr2)
print("division:",arr1/arr2)
```

```python
## universal function
arr=np.array([2,3,4,5,6])
## square root
print(np.sqrt(arr))

## Exponential
print(np.exp(arr))

##sign
print(np.sin(arr))

print(np.log(arr))
```

```python
## statistical concepts--normalization
## to have a mean of 0 and statement deviation of 1

data = np.array([1, 2, 3, 4, 5])

# Calculate the mean and standard deviation

mean = np.mean(data)

std_dev = np.std(data)

# Normalize the data

normalized_data = (data -mean) / std_dev

print("Normalized data:", normalized_data)

# Output:

# Normalized data: [-1.26491106 -0.63245553 0. 0.63245553 1.26491106]
```

```python
# Mean

mean= np.mean(data)

print("Mean:", mean)

# Median

median= np.median(data)

print("Median:", median)

#Standard deviation

std_dev = np.std(data)

print("Standard Deviation:", std_dev)

# Variance

variance =np.var(data)

print("Variance:",variance)
```
