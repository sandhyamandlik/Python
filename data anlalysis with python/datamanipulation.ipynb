```python
import pandas as pd
```

```python
df=pd.read_csv('data.csv')
df.head(5)
```

```python
df.describe()
```

```python
df.dtypes
```

```python
## handling missing values

df.isnull().any(axis=1)
```

```python
df.isnull().sum()
```

```python
df.fillna(0)
```

```python
## filling missing values with the mean of the columns
df['sales_fillNA']=df['sales'].fillna(df['sales'].mean())
df
```

```python
## renaming columns
df=df.rename(columns={'Date':'sales_Data'})
df.head()
```

```python
## change datatype

df['value_new']=df['value'].astype(float)
df.head()
```

```python
df['new value']=df['value'].apply(lambda x:x*2)
df.head()
```

```python
## data aggregating and grouping

df.head()
```

```python
grouped_mean=df.groupby('product')['value'].mean()
print(grouped_mean)
```

```python
grouped_sum=df.groupby('product','region')['value'].sum()
print(grouped_sum)
```

```python
## aggregating multiple functions
grouped_age=df.groupby('region')['value'].agg(['mean','sum','count'])
grouped_age
```

```python
## merging and joining dataframes

import pandas as pd

df1=pd.DataFrame({'key':['A','B','C'],'value1':[1,2,3]})
df2=pd.DataFrame({'key':['A','B','D'],'value2':[4,5,6]})
```

```python
df1
```

```python
df2
```

```python
## merge dataframes on the 'key columns'
pd.merge(df1,df2,on="key",how="inner")
```

```python
pd.merge(df1,df2,on="key",how="outer")
```

```python
pd.merge(df1,df2,on="key",how="left")
```
